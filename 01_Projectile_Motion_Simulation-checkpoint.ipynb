This project simulates the trajectory of a projectile launched with an initial velocity and angle using classical mechanics.

## Objective

The goal of this project is to model projectile motion under the influence of gravity and visualize the resulting trajectory.

## Physics Background

Projectile motion occurs when an object is launched with an initial velocity \(u\) at an angle \(theta\) with the horizontal. Assuming air resistance is negligible, the motion can be described using the following equations.

#### Horizontal Position

$$
x = u\cos(\theta)t
$$

#### Vertical Position

$$
y = u\sin(\theta)t - \frac{1}{2}gt^2
$$

where

$$
g = 9.8 \, \text{m/s}^2
$$

is the acceleration due to gravity.

#### Time of Flight

For a projectile that lands at the same vertical level from which it was launched,

$$
T = \frac{2u\sin(\theta)}{g}
$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# User Input
u = float(input("Enter initial velocity u (m/s): "))
theta = float(input("Enter the angle: "))
angle_unit = int(input("Choose 1. Radian or 2. Degree: "))

# Angle Conversion
if angle_unit == 1:
    theta_radians = theta

elif angle_unit == 2:
    theta_radians = math.radians(theta)

else:
    print("Invalid entry!")
    exit()

# Trigonometric Values
cos_value = math.cos(theta_radians)
sin_value = math.sin(theta_radians)

# Gravity
g = 9.8

# Flight Time
T = (2 * u * sin_value) / g

# Generate 200 time points from launch to landing
t = np.linspace(0, T, 200)

# Lists for coordinates
axis_x = []
axis_y = []

# Calculate trajectory
for time in t:
    x = u * cos_value * time
    y = u * sin_value * time - 0.5 * g * time**2

    axis_x.append(x)
    axis_y.append(y))

# Plot
plt.plot(axis_x, axis_y, color='m', label='trajectory')

plt.title("Projectile Motion Simulation")
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Height (m)")
plt.grid(True)
plt.legend()

plt.show()
```

## Results

The trajectory follows a parabolic path as predicted by classical mechanics.

## Conclusion

The simulation successfully reproduced the expected behavior of projectile motion and demonstrated the application of Python and numerical methods to solve physics problems.
