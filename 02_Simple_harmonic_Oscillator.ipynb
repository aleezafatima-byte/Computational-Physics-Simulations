## Theory

### Definition

A **Simple Harmonic Oscillator (SHO)** is a system that undergoes **Simple Harmonic Motion (SHM)**, in which the restoring force is directly proportional to the displacement from the equilibrium position and acts in the opposite direction.

### Hooke's law

The general restoring force is

$$
F_r=-kx
$$

where,

- $F_r$ = Restoring force
- $k$ = Spring constant (for a spring-mass system)
- $x$ = Displacement from equilibrium

> **Note:** The negative sign indicates that the restoring force always acts opposite to the displacement, pulling the object back toward the equilibrium position.



**Example**

A *mass attached to an ideal spring* is the most common example of a Simple Harmonic Oscillator.

When the mass is displaced from its equilibrium position and released, it oscillates back and forth due to the restoring force provided by the spring.


### Prerequisite: Angular Frequency
The angular frequency is

$$
\omega=\sqrt{\frac{k}{m}}
$$

where,

- $\omega$ = Angular frequency (rad/s)
- $k$ = Spring constant (N/m)
- $m$ = Mass (kg)


## Derivation

### 1. Differential Equation of Motion

Using **Newton's Second Law**

$$
F=ma
$$

and **Hooke's Law**

$$
F=-kx,
$$

we obtain

$$
m\frac{d^2x}{dt^2}=-kx.
$$

Dividing both sides by $m$,

$$
\frac{d^2x}{dt^2}+\frac{k}{m}x=0.
$$

Since

$$
\omega=\sqrt{\frac{k}{m}},
$$

the governing equation becomes

$$
\boxed{\frac{d^2x}{dt^2}+\omega^2x=0.}
$$

This is the **differential equation of the Simple Harmonic Oscillator**.


## 2. Analytical solution of differential equation: Position as a Function of Time

The analytical solution of the differential equation is

$$
\frac{d^2x}{dt^2} + \omega^2x = 0
$$

$$
\frac{d^2x}{dt^2} = -\omega^2x
$$

Multiply both sides by $$\frac{2dx}{dt}$$

$$
\frac{2dx}{dt}* \frac{d^2x}{dt^2} = -\omega^2x * 2\frac{dx}{dt}
$$

Now integrate both sides you will get-

$$
(\frac{dx}{dt})^2= -\omega^2x^2 + c
$$

where,

c= constant of integration

Now solve it like IVP: 

$at x=A$       $\frac{dx}{dt}=v=0$


Put this in equation we got

$$
0= -\omega^2A^2 + c
$$

$$
c=\omega^2A^2
$$

Put c in the equation, we get

$$
(\frac{dx}{dt})^2 = -\omega^2x^2 + \omega^2A^2
$$

$$
(\frac{dx}{dt})^2 = \omega^2(A^2 - x^2)
$$

$$
\frac{dx}{dt} = \omega\sqrt{A^2 - x^2}
$$

Now integrate

$$
\int \frac{dx} {\sqrt{A^2-x^2}} = \int\omega*dt
$$

$$
sin^-1\frac{x}{A}= \omega*t + \phi
$$

$$
\frac{x}{A}=\sin(\omega t+\phi)
$$

Therefore we get

$$
x(t)=A\sin(\omega t+\phi),
$$

or equivalently

$$
x(t)=A\cos(\omega t+\phi)
$$

where,

- $A$ = Amplitude
- $\omega$ = Angular frequency
- $\phi$ = Phase constant



The following code computes the analytical displacement and plots it as a function of time.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
h = 0.01
t0 = 0
tf = 10

A = 5
k = 9.0
m = 1.0

omega = np.sqrt(k/m)

# Time array
t = np.arange(t0, tf + h, h)

# Analytical solution
x = A * np.cos(omega * t)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, x,linewidth=2,label="Analytical solution")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Simple Harmonic Motion")
plt.grid(True)
plt.legend(loc="upper right")

plt.savefig("SHM_simulation1.png",dpi=300)
plt.show()
```

## 3. Equation of Velocity

From the acceleration equation,

$$
a=-\omega^2x
$$

Using

$$
a=\frac{dv}{dt},
$$

and applying the chain rule,

$$
\frac{dv}{dt}
=
\frac{dv}{dx}\frac{dx}{dt}
=
v\frac{dv}{dx},
$$

we get

$$
v\frac{dv}{dx}
=
-\omega^2x.
$$

Rearranging,

$$
v\,dv
=
-\omega^2x\,dx.
$$

Integrating both sides,

$$
\int_0^v v\,dv
=
-\omega^2
\int_A^x x\,dx.
$$

Evaluating,

$$
\left[\frac{v^2}{2}\right]_0^v
=
-\omega^2
\left[\frac{x^2}{2}\right]_A^x,
$$

which gives

$$
\frac{v^2}{2}
=
-\frac{\omega^2}{2}(x^2-A^2).
$$

Therefore,

$$
v^2
=
\omega^2(A^2-x^2).
$$

Taking the square root,

$$
\boxed{v=\pm\omega\sqrt{A^2-x^2}}
$$

The positive sign represents motion in the positive direction, while the negative sign represents motion in the opposite direction.


The following code computes the analytical velocity of the simple harmonic oscillator and plots its variation as a function of time.

```python
#Velocity vs Time
v = -A * omega * np.sin(omega * t)

#Plot
plt.plot(t, v)
plt.xlabel("Time(s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs Time")
plt.grid(True)

plt.savefig("SHM_simulation2.png",dpi=300)
plt.show()
```

## 4. Equation of Acceleration

Given Hooke's Law,

$$
F = -kx
$$

Using Newton's Second Law,

$$
F = ma
$$

Therefore,

$$
ma = -kx
$$

Dividing both sides by \(m\),

$$
a = -\frac{k}{m}x
$$

Since the angular frequency is

$$
\omega = \sqrt{\frac{k}{m}},
$$

therefore,

$$
\omega^2 = \frac{k}{m}.
$$

Substituting into the acceleration equation,

$$
a = -\omega^2 x
$$

Hence,

$$
\boxed{a = -\omega^2 x}
$$

This is the acceleration equation of a **Simple Harmonic Oscillator (SHO)**.

The following code computes the analytical acceleration of the simple harmonic oscillator and plots its variation as a function of time.

```python
# Acceleration vs Time
a = -omega**2*x

plt.plot(t,a)
plt.xlabel("time (s)")
plt.ylabel("acceleration (m/s^2")
plt.title("acceleration vs Time")
plt.grid(True)

plt.savefig("SHM_simulation4.png",dpi=300)
plt.show()
```

## 5. Relation Between Velocity and Displacement (Phase Space)

From the velocity equation,

$$
v=\omega\sqrt{A^2-x^2}
$$

Squaring both sides,

$$
v^2=\omega^2(A^2-x^2)
$$

Rearranging,

$$
\frac{v^2}{\omega^2}+x^2=A^2
$$

Dividing both sides by $A^2$,

$$
\frac{v^2}{\omega^2A^2}+\frac{x^2}{A^2}=1
$$

This is the standard equation of an ellipse.

The standard equation of an ellipse is

$$
\frac{x^2}{a^2}+\frac{y^2}{b^2}=1
$$

Comparing the two equations,

$$
\frac{x^2}{A^2}+\frac{v^2}{(\omega A)^2}=1
$$

we obtain

- Semi-major axis:

$$
a=A
$$

- Semi-minor axis:

$$
b=\omega A
$$

Hence, the graph between displacement $(x)$ and velocity $(v)$ is an **ellipse**, known as the **phase space trajectory** of Simple Harmonic Motion.

The maximum values are

$$
x=\pm A
$$

and

$$
v=\pm\omega A.
$$

The following code computes the analytical relationship between displacement and velocity and plots the phase-space trajectory of the simple harmonic oscillator.

```python
#Phase Space Plot
plt.figure(figsize=(6,6))
plt.plot(x, v)
plt.xlabel("Displacement (m)")
plt.ylabel("Velocity (m/s)")
plt.title("Phase Space of SHM")
plt.grid(True)

plt.savefig("SHM_simulation3.png",dpi=300)
plt.show()
```

## 6. Energy of the Oscillator

The total mechanical energy remains constant.

### Potential Energy

$$
PE=\frac12kx^2
$$

### Kinetic Energy

$$
KE=\frac12mv^2
$$

Putting value of v

$$
KE=\frac12m\omega^2(A^2-x^2)
$$

### Total Energy

$$
E=KE+PE=\frac12kA^2
$$

Although kinetic and potential energy continuously interchange during the motion, the total mechanical energy remains constant for an ideal (undamped) oscillator.


The following code computes the kinetic, potential, and total mechanical energies of the simple harmonic oscillator and plots them as functions of displacement.

```python
# Energy vs Displacement

#Energy formulas-
PE = 0.5 * k * x**2                            #Potential energy
KE = 0.5 * m * omega**2 * (A**2 - x**2)        #Kinetic energy
TE = PE + KE                                   #Total energy
print("Total Energy =", 0.5*k*A**2)

plt.plot(x,PE,label="Potential energy")
plt.plot(x,KE,label="Kinetic energy")
plt.plot(x,TE,label="Total energy")
plt.xlabel("Displacement(m)")
plt.ylabel("Energy(J)")
plt.title("Energy vs Displacement of SHM")
plt.grid("True")
plt.legend(loc="upper right")

plt.savefig("SHM_simulation5.png",dpi=300)
plt.show()
```

The following code computes the kinetic, potential, and total mechanical energies of the simple harmonic oscillator and plots their variation as a function of time.

```python
# Energy vs Time

plt.plot(t, PE, label="Potential Energy")
plt.plot(t, KE, label="Kinetic Energy")
plt.plot(t, TE, label="Total Energy")
plt.xlabel("Time(s)")
plt.ylabel("Energy(J)")
plt.title("Energy vs Time of SHM")
plt.grid("True")
plt.legend(loc="upper right")

plt.savefig("SHM_simulation6.png",dpi=300)
plt.show()
```

## Conclusion

The analytical solution satisfies the equations of Simple Harmonic Motion.

- The phase-space trajectory forms an ellipse
- The total mechanical energy remains constant throughout the motion, confirming conservation of energy for an ideal oscillator.
